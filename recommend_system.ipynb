```python
ndf = df[[ 'uniq_id', 'product_name' , 'description' , 'product_information' , 'price' , 'amazon_category_and_sub_category' ,'manufacturer']]
```

```python
import re

def clean_text(text):
    if isinstance(text, str):
        # Remove everything except letters, numbers, and spaces
        return re.sub(r'[^\w\s]', '', text)
    return ""
```

```python
ndf['description'] = ndf['description'].apply(clean_text)
ndf['product_name'] = ndf['product_name'].apply(clean_text)
ndf['product_information'] = ndf['product_information'].apply(clean_text)
ndf['price'] = ndf['price'].apply(clean_text)
ndf['category_and_sub_category'] = ndf['category_and_sub_category'].apply(clean_text)
ndf['manufacturer'] = ndf['manufacturer'].apply(clean_text)
```

```python
def remove_spaces(text):
    if isinstance(text, str):
        return text.replace(" ", "")
    return ""
```

```python
ndf['manufacturer'] = ndf['manufacturer'].apply(remove_spaces) 
```

```python
ndf.head(2)
```

```python
ndf['Tags'] = ndf['description'] + ndf['product_information'] + ndf['category_and_sub_category']+ ndf['manufacturer']+ ndf['price']
```

```python
ndf
```

```python
new_df = ndf[['uniq_id','product_name','Tags']]
```

```python
new_df.head(2)
```

```python
new_df['Tags'][2]
```

```python
# lower case 
def lowercase(text):
    return text.lower()
```

```python
new_df['Tags'] = new_df['Tags'].apply(lowercase)
```

```python
# text vectorization using tdf 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(stop_words='english') # remove words like is ,the ,for ,etc.
tfidf_matrix = tfidf.fit_transform(new_df['Tags']) # a matrix where each row is a product and each column is a word
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations(product_index, cosine_sim=cosine_sim): # takes the product index 
    sim_scores = list(enumerate(cosine_sim[product_index])) # fetch the similarity and store  it in tuple form , (index:similarity)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)  # sort in decending order 
    sim_scores = sim_scores[1:6]  # top 5, excluding the product itself
    product_indices = [i[0] for i in sim_scores]  # fetch the actual product rows from the dataframe
    return ndf['product_name'].iloc[product_indices].values
```
